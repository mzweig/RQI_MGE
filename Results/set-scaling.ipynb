```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### Time in Hours ### 
arnoldi_part = np.array([3.06, 1.90, 1.84])
arnoldi_full = np.array([7.99, 4.04, 2.34])
rqi_nop = np.array([5.56, 2.86, 1.63])
rqi_precon = np.array([4.71, 2.46, 1.55])
cores = np.array([128, 256, 512])
```

```python
plt.loglog(cores, arnoldi_part, label="Arnoldi, Partition Upscatter")
plt.loglog(cores, arnoldi_full, label="Arnoldi, Full Partition")
plt.loglog(cores, rqi_nop, label="RQI, No Preconditioner")
plt.loglog(cores, rqi_precon, label="RQI, Preconditioner")
plt.xlabel("Number of Cores")
plt.ylabel("Run Time (hrs)")
plt.legend()
```

```python
# Print Slopes
print("Slopes: ")
slope, intercept = np.polyfit(np.log(cores), np.log(arnoldi_part), 1)
print("Arnoldi, Partition Upscatter: ")
print(slope)
slope, intercept = np.polyfit(np.log(cores), np.log(arnoldi_full), 1)
print("Arnoldi, Full Partition: ")
print(slope)
slope, intercept = np.polyfit(np.log(cores), np.log(rqi_nop), 1)
print("RQI, No Preconditioner: ")
print(slope)
slope, intercept = np.polyfit(np.log(cores), np.log(rqi_precon), 1)
print("RQI, Preconditioner")
print(slope)
```
